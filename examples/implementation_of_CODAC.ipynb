```python
import CODAC 

I1 = CODAC.Interpro('IPR000980', 'Homo sapiens')
uniprotlist = CODAC.Interpro.fetch_uniprotid(I1)
```

```python
(uniprotlist)
```

```python
CODAC.Uniprot.makeRefFile(uniprotlist, 'IPR000980')
```

```python
CODAC.Uniprot.make_fastafile('SH2', 'Domain_Referencer_IPR000980_HUMAN.csv', 'IPR000980', specie='HUMAN')
```

```python
dict_main = {}

for uid in uniprotlist:
    
    dict_main[uid] = {}
    
    I2 = CODAC.Uniprot(uid)
    dom = CODAC.Uniprot.Domains(I2)
    gene = CODAC.Uniprot.Gene(I2)
    refseq = CODAC.Uniprot.RefSeq(I2)
    domarch = CODAC.Uniprot.DomainArchitecture(I2)
    dict_main[uid]['domain'] = dom
    dict_main[uid]['gene'] = gene
    dict_main[uid]['refseq'] = refseq
    dict_main[uid]['domarch'] = domarch
    print(uid)
```

```python
for i in uniprotlist:
    print(dict_main[i]['domarch'])
```
